```python
import pandas
import numpy as np
data = pandas.read_csv('titanic.csv', index_col='PassengerId')
data[:10]
```

```python
data1 = data[['Pclass','Fare','Age','Sex','Survived']][np.invert(np.isnan(data['Age']))].reset_index(drop=True)
i = 0
while (i < data1.shape[0]):
    if (data1['Sex'][i] == 'male'):
        data1['Sex'][i] = 1
        #print('1')
    else:
        data1['Sex'][i] = 0
        #print('0')
    i = i + 1
print(data1)
```

```python
X = data1[['Pclass','Fare','Age','Sex']]
Y = data1[['Survived']]
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=241)
clf.fit(X, Y)
importances = clf.feature_importances_
clf.feature_importances_
```
