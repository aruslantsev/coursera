```python
import numpy as np
import pandas as pd

data = pd.read_csv('salary-train.csv')
```

```python
data_train = data.replace('[^a-zA-Z0-9]', ' ', regex=True)
for i in range(data_train.shape[0]):
    data_train['FullDescription'][i]=data_train['FullDescription'][i].lower()
    data_train['LocationNormalized'][i]=data_train['LocationNormalized'][i].lower()
    if (i % 500 == 0):
        print(i)
    #data_train['ContractTime'][i]=data_train['ContractTime'][i].lower()
data_train[:10]
```

```python
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=5)
X_train = tfv.fit_transform(data_train['FullDescription'])
```
