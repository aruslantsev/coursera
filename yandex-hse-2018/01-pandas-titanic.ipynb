```python
import pandas

data = pandas.read_csv('titanic.csv', index_col='PassengerId')
print(data.head())
sex = data['Sex'].value_counts()
print ('Sex: male:', sex['male'],', female:', sex['female'])
survived = data['Survived'].value_counts()
print ('Survived:', "%.2f" % (survived[1]/survived.sum()*100))
pclass = data['Pclass'].value_counts()
print ('1st class:', "{:0.2f}".format(pclass[1]/pclass.sum()*100))
print ('Mean:', "{:0.2f}".format(data['Age'].mean()), ', Median:', "{:0.2f}".format(data['Age'].median()))
print ('Pearson:', "{:0.2f}".format(data['SibSp'].corr(data['Parch'], method='pearson')))

def get_name(name):
    name=name.replace('Mrs','')
    name=name.replace('Miss','')
    name=name.replace("'", '')
    name=name.replace('"','')
    name=name.replace(')','')
    name=name.replace('.','')
    if (name.find('(') != -1):
        name=name.split('(')[1]
    else:
        name=name.split(',')[1]
    name=name.lstrip()
    name=name.split(' ')[0]    
        
    return name
    
w_names=data[data['Sex']=='female']['Name'].map(get_name)
n_counts=w_names.value_counts()
print(n_counts.head(1).index[0])

# Sample
print(data.reset_index().groupby(['Sex', 'Survived'])['PassengerId'].count(), '\n', 
      data.reset_index().groupby(['Sex', 'Survived']).count())
```
