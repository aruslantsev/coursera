```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
X = np.linspace(start=1, stop=15, num=100)
Y = f(X)
plt.plot(X, Y)
```

```python
part_x = []
part_x.append(np.array([1., 15.]))
part_x.append(np.array([1., 8., 15.]))
part_x.append(np.array([1., 4., 10., 15.]))
```

```python
A = []
b = []
for x in part_x:
    A.append(np.hstack([x.reshape(-1, 1)**i for i in range(len(x))]))
    print x.shape
    b.append(f(x))
```

```python
C = []
for i in range(len(A)):
    C.append(np.linalg.inv(A[i]).dot(b[i]))
```

```python
y_iter = []
for i in range(len(C)):
    y = np.zeros(len(X))
    for j in range(len(C[i])):
        y += C[i][j] * X**j
    y_iter.append(y)
```

```python
for i in range(len(C)):
    plt.plot(X, y_iter[i], X, Y)
```
